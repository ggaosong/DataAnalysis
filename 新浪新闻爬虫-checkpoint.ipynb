```python
import requests
from bs4 import BeautifulSoup
def getNewsDetail(newsurl):
    result = {}
    res = requests.get(newsurl)
    res.encoding = 'utf-8'
    soup = BeautifulSoup(res.text,'html.parser')
    result['total'] = soup.select('.main-title')[0].text
    result['newssource'] = soup.select('#article p')[:-1]
    timesource =  soup.select('.date-source')[0].contents[1].text
    result['dt'] = datetime.strptime(timesource,'%Y年%m月%d日 %H:%M')
    result['eidtor'] = soup.select('.show_author')[0].text.strip('责任编辑：')
    result['comments'] = getcommentcounts(newsurl)
    return result
```

```python
def parseListLinks(url):
    newsdetails = []
    dl = requests.get(url)
    aa = dl.text.strip()[26:].strip()[:-14]
    lf = json.loads(aa)
    for ent in lf['result']['data']:
        newsdetails.append(getNewsDetail(ent['url']))
    return newsdetails
  
```

```python
url = 'https://feed.sina.com.cn/api/roll/get?pageid=121&lid=1356&num=20&versionNumber=1.2.4&page=2&encode=utf-8&callback=feedCardJsonpCallback&_=1537195983168'
parseListLinks(url)
```

```python
url = 'https://feed.sina.com.cn/api/roll/get?pageid=121&lid=1356&num=20&versionNumber=1.2.4&page={}&encode=utf-8&callback=feedCardJsonpCallback&_=1537195983168'
news_total = []
for i in range(1,4):
    newsurl = url.format(i)
    newsary = parseListLinks(newsurl)
    news_total.extend(newsary)
```

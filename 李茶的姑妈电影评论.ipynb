```python
import requests
from bs4 import BeautifulSoup
import re

def getcomments(url,i):
    
    
    result = {}
   
    res = requests.get(url)
    res.encoding = 'utf-8'
    result['source'] = soup.select('.mod_short')[i].contents[1].text
    

    ss = soup.select('.mod_short')[i].contents[3].text
    num = re.findall('\d+',ss)
    result['score'] = num[-2]
    

    
    

    return result 
```

```python
getcomments('http://movie.mtime.com/254620/reviews/short/new-2.html',0)
```

```python
def getone(URL):

    cone = []

    for i in range(0,26):
        gg = getcomments(URL,i)
        cone.append(gg)
    return cone

cc = getone('http://movie.mtime.com/254620/reviews/short/new-2.html')
print(cc)
      
```

```python
import pandas as pd


ff = pd.read_csv('FilmComments.csv')
word = ff['source']
lc = str(word)
```

```python
import jieba.analyse
from PIL import Image,ImageSequence
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud,ImageColorGenerator

result=jieba.analyse.textrank(lc,topK=200,withWeight=True,allowPOS=('a','e','n','nr','ns'))
keywords = dict()
for i in result:
    keywords[i[0]]=i[1]
print(keywords)
```

```python

image= Image.open('E:\Pythondata\hb.jpg')
graph = np.array(image)
wc = WordCloud(font_path='./fonts/simhei.ttf',background_color='White',max_words=100,mask = graph)
wc.generate_from_frequencies(keywords)
image_color = ImageColorGenerator(graph)
plt.imshow(wc)
plt.imshow(wc.recolor(color_func=image_color))
plt.axis("off")
plt.show()
wc.to_file('dream.png')
```
